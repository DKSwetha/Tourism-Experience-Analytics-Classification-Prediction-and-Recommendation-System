# **Exploratory Data Analysis-**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

master_df = pd.read_csv("../data/processed/master_cleaned.csv")

master_df.shape
```

User Distribution Across Geography

```python
# Users per Continent
plt.figure()
master_df['Continent'].value_counts().plot(kind='bar')
plt.title("User Distribution by Continent")
plt.xlabel("Continent")
plt.ylabel("Number of Visits")
plt.show()
```

```python
# Users per Country (Top 10)
plt.figure()
master_df['Country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries by Visits")
plt.xlabel("Country")
plt.ylabel("Number of Visits")
plt.show()
```

```python
# Users per Region
plt.figure()
master_df['Region'].value_counts().plot(kind='bar')
plt.title("User Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Number of Visits")
plt.show()
```

Attraction Types & Popularity

```python
# Attraction Type Popularity
plt.figure()
master_df['AttractionType'].value_counts().plot(kind='bar')
plt.title("Attraction Type Popularity")
plt.xlabel("Attraction Type")
plt.ylabel("Number of Visits")
plt.show()
```

```python
# Average Rating per Attraction Type
plt.figure()
master_df.groupby('AttractionType')['Rating'].mean().sort_values().plot(kind='bar')
plt.title("Average Rating by Attraction Type")
plt.xlabel("Attraction Type")
plt.ylabel("Average Rating")
plt.show()
```

VisitMode vs Demographics

```python
# VisitMode Distribution
plt.figure()
master_df['VisitMode'].value_counts().plot(kind='bar')
plt.title("Visit Mode Distribution")
plt.xlabel("Visit Mode")
plt.ylabel("Count")
plt.show()
```

```python
# VisitMode vs Continent
pd.crosstab(master_df['Continent'], master_df['VisitMode']).plot(kind='bar')
plt.title("Visit Mode by Continent")
plt.xlabel("Continent")
plt.ylabel("Count")
plt.show()
```

Rating Distribution

```python
# Overall Rating Distribution
plt.figure()
master_df['Rating'].value_counts().sort_index().plot(kind='bar')
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
# Average Rating by Region
plt.figure()
master_df.groupby('Region')['Rating'].mean().plot(kind='bar')
plt.title("Average Rating by Region")
plt.xlabel("Region")
plt.ylabel("Average Rating")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(master_df[['Rating','VisitMonth','VisitYear']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
